**FUNCION LINEAL:**

La funcion lineal es la **contiene una constante**, usaremos el siguiente **caso para explicar:**

Un grupo de estudiantes **cuelga varios objetos de distinto peso**, cada uno en un resorte, los estudiantes m**iden cuanto se extiende el resorte** a partir del largo natural. Los resultados son los de la siguiente tabla:

$$\frac{Peso (kg)|1.6 | 1.92 | 2.16 | 2.56}
{Def.(cm)|2.0  |2.40  |2.70  |3.20}$$
Al graficar la tabla, se nota una regularidad:
Screenshot 2024-09-05 115937.png
Podemos darnos cuenta que **a medida que sube el peso**, la deformacion del resorte **incrementa de forma constante**

La **formula para determinar la constante (o pendiente)** es la siguiente:
$$Pendiente = \frac{Diferencia Vertical}{Diferencia Horizontal}$$

En este caso, **la diferencia vertical** (cm) es: $$2.4 - 2 = 0.4$$ $$2.7 - 2 = 0.7$$ $$3.2 - 2 = 1.2$$

Y la **diferencia horizontal** (kg) es: $$1.92 - 1.6 = 0,32$$ $$2.16 - 1.6 = 0.56$$  $$2.56 - 1.6 = 0.96$$

# GUIA 1, PROBLEMA 3:

La latencia de una red corresponde al tiempo que tarda un paquete de datos en viajar desde el punto de origen al destino. En otras palabras, es el tiempo que transcurre desde que se envía una solicitud de un dispositivo hasta que se recibe una respuesta del servidor u otro dispositivo.

La latencia puede verse afectada por la velocidad de transmisión de los datos a través de los cables o conexiones inalámbricas, la distancia física entre los dispositivos, el tiempo que tarda un computador en procesar o reenviar los datos, etc.

En redes de comunicación, como las videoconferencias, juegos en línea y transmisión de datos en tiempo real, es deseable una baja latencia, de lo contrario pueden existir retrasos en la comunicación y afectar negativamente la experiencia del usuario.

Según el contexto mencionado, crea una función en *Python* que permita calcular la **latencia real** de una red dada una latencia estimada. Suponga que la latencia real es un 20% mayor a la latencia estimada.

calcule la latencia real para una latencia estimada de:

a)   200 milisegundos.

b)   149 milisegundos.

c)   74 milisegundos.

```python
def lat_real(lat_estimada):
    return lat_estimada + (lat_estimada * 0.20)
print(f'Si se tiene una latencia estimada de 200 milisegundos, la latencia real seria de {lat_real(200)} milisegundos')
print(f'Si se tiene una latencia estimada de 149 milisegundos, la latencia real seria de {lat_real(149)} milisegundos')
print(f'Si se tiene una latencia estimada de 74 milisegundos, la latencia real seria de {lat_real(74)} milisegundos')
```

```python
#7.2
import numpy as np

def U(t):
  return 1000 / (1 + 9 * np.exp(-0.5 * t))
print(f'Transcurridos 12 meses, la red social tendra {U(12):.0f} usuarios')
```
