3. Calcula el tiempo de carga y descarga para un archivo de $750$ MB.

```python
#1:
import numpy as np
from scipy.optimize import fsolve

size = np.array([50, 100, 250, 500, 1000])
carga = np.array([10, 20, 50, 100, 200])
descarga = np.array([8, 16, 40, 80, 160])

m_carga, n_carga = np.polyfit(size, carga, 1)
m_descarga, n_descarga = np.polyfit(size, descarga, 1)

print(f'La pendiente de la función "carga" es {(m_carga):.2f}, y el coeficiente de posición es {(n_carga):.2f}')
print(f'La pendiente de la función "descarga" es {(m_descarga):.2f}, y el coeficiente de posición es {(n_descarga):.2f}')
```

```python
#3:
import numpy as np
from scipy.optimize import fsolve

size = np.array([50, 100, 250, 500, 1000])
carga = np.array([10, 20, 50, 100, 200])
descarga = np.array([8, 16, 40, 80, 160])

m_carga, n_carga = np.polyfit(size, carga, 1)
m_descarga, n_descarga = np.polyfit(size, descarga, 1)

print(f'El tiempo de carga para un archivo de 750 MB es {(m_carga*750+n_carga):.2f} segundos')
print(f'El tiempo de carga para un archivo de 750 MB es {(m_descarga*750+n_descarga):.2f} segundos')
```

```python
#4:
import numpy as np
from scipy.optimize import fsolve

def carga(x):
  return 0.05*x-163

x = np.linspace(0,100,1)

solucion = fsolve(carga,x)
print(f'Para un tiempo de carga de 163 segundos, el tamaño del archivo seria de {(solucion[0]):.2f} MB')
```

```python
#5:
import numpy as np
from scipy.optimize import fsolve

def descarga(x):
  return 0.16*x-195

x = np.linspace(0,100,1)

solucion = fsolve(descarga,x)

print(f'Para un tiempo de descarga de 195 segundos, el tamaño del archivo seria de {(solucion[0]):.2f} MB, por ende, el trabajador esta equivocado.')
```

$$V(x) = 300x + 500$$

```python
#4:
def V(x):
  return 300*x + 500

print(f'El numero de visitas diarias al ser ejecutadas 5 campañas publicitarias sería de {V(5)} personas')
```

```python
#5:
import numpy as np
from scipy.optimize import fsolve

def V(x):
  return 300*x + 500 - 4700
x = np.linspace(0,100,1)
solucion = fsolve(V,x)
print(f'Para alcanzar 4700 visitas diarias, se deben ejecutar {solucion[0]:.0f} campañas publicitarias')
```
