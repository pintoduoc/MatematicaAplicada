**FUNCION LINEAL:**

La funcion lineal es la **contiene una constante**, usaremos el siguiente **caso para explicar:**

Un grupo de estudiantes **cuelga varios objetos de distinto peso**, cada uno en un resorte, los estudiantes m**iden cuanto se extiende el resorte** a partir del largo natural. Los resultados son los de la siguiente tabla:

$$\frac{Peso (kg)|1.6 | 1.92 | 2.16 | 2.56}
{Def.(cm)|2.0  |2.40  |2.70  |3.20}$$
Al graficar la tabla, se nota una regularidad:

Podemos darnos cuenta que **a medida que sube el peso**, la deformacion del resorte **incrementa de forma constante**

La **formula para determinar la constante (o pendiente)** es la siguiente:
$$Pendiente = \frac{Diferencia Vertical}{Diferencia Horizontal}$$

En este caso, **la diferencia vertical** (cm) es: $$2.4 - 2 = 0.4$$ $$2.7 - 2 = 0.7$$ $$3.2 - 2 = 1.2$$

# **SUCESIONES**

Una sucecion es una serie de terminos $a_{n}$, donde cada uno de ellos esta asociado a un **valor $n$**, el cual es un numero **natural** que nos **indica el lugar que ocupa el termino**, por ejemplo:

$a_{n} = a_{1}, a_{2}, a_{3}, a_{4}, ...$

donde $a_{n}$ representa el termino en el lugar $n$ de la sucesion, con $n = 1, 2, 3,...$

Nombre de la sucesion $(b)$ --> $b_{k}$<-- Sub indice $(k)$

El nombre de la sucecion es el numero que usamos para identificarla, mientras que el numero subindice es el lugar que ocupa en la sucecion.

**Ejemplo de codigo:** 

Hacer una lista con los primeros 4 numeros de la sucesion $a_{n} = 3n$, imprimirlos todos, tambien imprimir el numero en la posicion $3$ y encontrar la posicion del numero $6$ en la sucesion

```python
lista = []
for i in range(1,5): #aqui ponemos un 1 y hasta donde queremos que agregue a la lista y se le suma 1
  lista.append(3*i) #Aqui ponemos la forma de la sucesion

print(f'En la sucesion, el numero en la posicion 3 es equivalente a: {lista[2]}')
print(f'Los primeros 4 numeros de la sucesion "3n" son: {lista}')

#Para saber en que posicion se encuentra un numero, usamos .index(numero), y al imprimirlo le sumamos uno
#Ejemplo para saber la posicion del numero 6:
print(f'La posicion en la que se encuentra el numero "6" es: {lista.index(6)+1}')
```

### **Sucesion aritmetica**

Son las que **tienen una diferencia constante**, por ejemplo:

En la susecion $4, 16, 28, 40, 52,...$, podemos darnos cuenta que es aritmetica pues la diferencia entre cada elemento es $12$.

Otro ejemplo seria la sucesion $500, 480, 460, 440, 420,...$, la cual tambien es aritmetica pues la diferencia entre cada elemento es de $-20$

La **diferencia** de una sucesion aritmetica se obtiene de la siguiente manera:
$$a_{n} - a-{n-1}$$
Ejemplo: Si $n = 4$, entonces:
$$d = a_{4} - a_{4-1} = a_{4} - a_{3}$$

Podemos calcular los terminos de una sucesion aritmetica de dos maneras:

1. Directamente, usando la formula para el termino de lugar $n$:

  $a_{n} = a_{1}+ (n+1)d$

2. Usando ciclos for desde i = 2 hasta $i = n$ teniendo en cuenta que:

  $a_{i} = a_{i-1} + d$
